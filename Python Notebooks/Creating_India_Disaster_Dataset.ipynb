```python
# Importing Required Packages

import numpy as np
import pandas as pd
import sqlite3
```

```python
# Read the 'FourDisaster_SouthAsianData2000to2020.csv' as disaster_df

disaster_df = pd.read_csv('FourDisaster_SouthAsianData2000to2020.csv')
disaster_df.head()
```

```python
# Shape of the data frame

disaster_df.shape
```

```python
# Extract the Year of every row and create a column 'Year'

# Define a function as extract the values

def get_year(entry):
    return entry.split('-')[0]

# Creating the Column 'Year'

disaster_df['Year'] = disaster_df['DisNo.'].apply(get_year)
disaster_df.shape
```

```python
# Connecting the data base with sqlite3

conn = sqlite3.connect(':memory:')
disaster_df.to_sql('DISASTER', conn, index = False)
```

```python
# Get the Column Names of the data frame

disaster_df.columns
```

```python
# Running SQL query to select Indian Regions with some specified information

query = 'SELECT Year,Country,ISO,"Disaster Type","Disaster Subtype",Location,"Total Deaths","No. Injured","No. Affected" \
"No. Homeless","Total Affected" FROM DISASTER WHERE Country = "India";'

result = pd.read_sql_query(query, conn)
print('The Shape of the Result Data frame is:', result.shape)
result.head()
```

```python
# Export the Data Frame as 'India_Disaster.csv'

file_path = 'India_Disaster.csv'
result.to_csv(file_path, index = False)
```
