```python
# Importing Packages

import pandas as pd
import numpy as np
```

```python
# Reading the India_Drought.csv data as drought_df

drought_df = pd.read_csv('India_Drought.csv')
drought_df.head()
```

```python
# Changing the Column 'Location' in list format

location_list = drought_df['Location'].tolist()
len(location_list)
```

```python
# Joining the elements of the location_list

join_location_list = ''.join(location_list)
len(join_location_list)
```

```python
# Find Unique Regions and their counts

from collections import Counter

# Split the string by commas to get a list of words
words_list = join_location_list.split(',')

# Count the occurrences of each unique word
word_counts = {}
for word in words_list:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Create a DataFrame from the result
drought_region = pd.DataFrame(list(word_counts.items()), columns=['StateName', 'Frequency'])

print(drought_region)
```
