### We observe that each of the 4 data frames have same number of columns and explicitely same columns

* I join the 4 data frames to form a full complete data frame
* Store the complete data in an object called finaldata
* I name the complete data frame as 'FourDisaster_SouthAsianData2000to2024.csv'

```python
# Concatinate the four data frame objects and check the shape

final_data = pd.concat([storm_data, flood_data, earth_quake_data,drought_data], axis = 0)

# Check the Shape of the data frame objcet final_data
print('Number of Rows in final_data are:', final_data.shape[0])
print('Number of Columns in final_data are:', final_data.shape[1])
```
