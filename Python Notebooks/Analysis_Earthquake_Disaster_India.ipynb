```python
# Importing Required Packages

import numpy as np
import pandas as pd
```

```python
# Reading the 'India_Earthquake.csv' data as earth_df

earth_df = pd.read_csv('India_Earthquake.csv')
earth_df.head()
```

```python
# Converting the Location column into list format

location_list = earth_df['Location'].tolist()
len(location_list)
```

```python
# Joining the Elements of the list - location_list

joined_location_list = ''.join(location_list)
len(joined_location_list)
```

```python
# Find Unique Regions and their counts in a data frame format

from collections import Counter

# Split the string by commas to get a list of words
words_list = joined_location_list.split(',')

# Count the occurrences of each unique word
word_counts = {}
for word in words_list:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Create a DataFrame from the result
earthquake_region = pd.DataFrame(list(word_counts.items()), columns=['StateName', 'Frequency'])

# Get the data frame
earthquake_region
```
